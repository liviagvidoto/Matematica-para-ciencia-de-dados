```python
#criando uma função simples
def f(x):
    return 2*x**2 +3*x + 1
print (f(2))
```

```python
#Utilizando uma lista
def polinomio(x, coeficientes):
    resultado = 0
    grau = len(coeficientes) - 1
    
    for i, coef in enumerate(coeficientes):
        resultado += coef * x**(grau - i)
    
    return resultado

coef = [2, 3, 1]
print(polinomio(2, coef))
```

```python
#usando Numpy
import numpy as np

coef = [2, 3, 1]
p = np.poly1d(coef)
print(p(2))
```
